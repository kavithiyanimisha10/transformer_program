```python
print("Welcome to the data analyzer and transformer program.")

lst=[]
def enter_data():
    """This function is used for enter data from the user.
    It will store the data in the form of list dataset."""

    choice=int(input("enter 1 for 1d list or 2 for 2d list :"))

    if choice==1:
        num=int(input("enter number of element:"))
        
        for i in range(1,num+1):
            ele=int(input(f"enter element {i}.:"))
            lst.append(ele)
    
    elif choice==2:
        
        row=int(input("enter number of rows:"))
        col=int(input("enter number of columns:"))
    
        for i in range(1,row+1):
            rows=[]
            for j in range(1,col+1):
                column=int(input(f"enter element {i},{j}:"))
                rows.append(column)
            lst.append(rows)
            
    print("Your data :",lst)
    print("\nYour data entered successfully.")

def data_summary():
    """This function is used for various statistical calculations.
    ex.,Total number of elements , sum of all elements , 
    minimum and maximum number of the list, average number of the list."""

    def total_element():
        total_ele=len(lst)
        print("Total elements are:",total_ele)

    def sum_num():
        if all(isinstance(i,list)for i in lst):
            print("sum:",sum(sum(i)for i in lst))
        else:
            print("sum:",sum(lst))

    def min_num():
        if all(isinstance(i,list)for i in lst):
            print("minimum number:",min(min(i)for i in lst))
        else:
            print("minimum number:",min(lst))

    def max_num():
        if all(isinstance(i,list)for i in lst):
            print("maximum number:",max(max(i)for i in lst))
        else:
            print("maximum number:",max(lst))

    def avg_num():
        if all(isinstance(i,list)for i in lst):
            avg=sum(sum(i)for i in lst)/len(lst)
            print("average of all numbers:",avg)
        else:
            print("average of all numbers:",sum(lst)/len(lst))

    total_element()
    sum_num()
    min_num()
    max_num()
    avg_num()

def factorial(n):
    """This function is used to find factorisation of any number."""
    f=1
    for i in range(2,n+1):
        f=f*i
    return f

def filter_data():
    """This function is used to filter our data.
    User has to enter 'threshold value' for filtering data."""
    
    if all(isinstance(i,list) for i in lst):
        data = [j for i in lst for j in i]
    else:
        data = lst

    threshold = int(input("Enter threshold value: "))
    filtered = list(filter(lambda x: x > threshold, data))
    print("Filtered Data:", filtered)

def sort_data():
    """This function is used to set the data in ascending or descending order."""
    
    if all(isinstance(i, list)for i in lst):
        print("Sorting each row:")
        for i in lst:
            i.sort()
            print(i)
    else:
        choice = int(input("1. Ascending  2. Descending : "))
        if choice == 1:
            print(sorted(lst))
        else:
            print(sorted(lst, reverse=True))

def return_dataset():
    """This function is used to return multiple values of the data.
    ex.,minimum and maximum value,average value."""
    
    if all(isinstance(i,list)for i in lst):
        data = [j for i in lst for j in i]
    else:
        data = lst

    return min(data), max(data), sum(data) / len(data)

while True:
    print("\nMAIN MENU.")
    print("1.Enter data.")
    print("2.Display Data Summary")
    print("3.Calculate factorial(Recurtion)")
    print("4.Filtered data by threshold(Lambda Function)")
    print("5.Sort data")
    print("6.Return dataset statistics")
    print("7.Exit")

    n=int(input("\nenter your choice:"))

    if n==1:
        print(enter_data.__doc__)
        enter_data()

    if n==2:
        print(data_summary.__doc__)
        data_summary()

    if n==3:
        print(factorial.__doc__)
        num=int(input("Enter number for foctorisation:"))
        print(f"factorial of {num} is :",factorial(num))

    if n==4:
        print(filter_data.__doc__)
        filter_data()

    if n==5:
        print(sort_data.__doc__)
        sort_data()

    if n==6:
        print(return_dataset.__doc__)
        mn, mx, avg = return_dataset()
        print("Min:", mn)
        print("Max:", mx)
        print("Average:", avg)
        
    if n==7:
        print("Thank you for using this transformer program.")
        print("Have a good day.")
        break
```
